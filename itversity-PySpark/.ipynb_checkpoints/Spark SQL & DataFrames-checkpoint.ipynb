```python
orderItems = sc.textFile('/Users/shreyas_rl/Desktop/git/itversity-data/retail_db/order_items/part-00000')
```

```python
orderItems.take(5)
```

```python
# filter operation in PySpark : normal python filter is static function, PySpark filter is instance function
```

```python
filteredOrderItems = orderItems.filter(lambda x : int(x.split(',')[1]) == 2)
filteredOrderItems.collect()
```

```python
mappedOrderItems = filteredOrderItems.map(lambda x : float(x.split(',')[4]))
mappedOrderItems.collect()
```

```python
reducedOrderItems = mappedOrderItems.reduce(lambda x,y : x+y)
reducedOrderItems
```
