```python
#SHUFFLE OPERATIONS : Each RDD record must be of the type : (K,V)  ---> Tuple(Key,Value)

myRDD = sc.textFile('/Users/shreyas_rl/Desktop/git/itversity-data/retail_db/order_items')
myRDD.take(5)
```

```python
#Get OrderID & Subtotal

mapRDD = myRDD.map(lambda x : func(x))
def func(x):
    temp = x.split(',')
    return((int(temp[1]),float(temp[4]))) # All records of an RDD must be tuples!!!

mapRDD.take(5) # This is called as a PairedRDD
```

```python
''' [(K1,[V1k1, V2k1, V3k1, ......, Vnk1]),
     (K2,[V1K2, V2K2, V3K2, ......, VnK2]),
     (K3,[V1K3, V2K3, V3K3, ......, VnK3]),
     ....................................]'''

groupedRDD = mapRDD.groupByKey()
groupedRDD.take(5)
```

```python
print(groupedRDD.count(),  mapRDD.count())
```

```python
orderRevenue = groupedRDD.map(lambda x : (x[0],sum(x[1])))
orderRevenue.take(5)
```

```python
#ReduceByKey = GroupByKey + Operation(Map):

orderRevenue = mapRDD.reduceByKey(lambda x,y : x+y)
orderRevenue.take(5)
```
