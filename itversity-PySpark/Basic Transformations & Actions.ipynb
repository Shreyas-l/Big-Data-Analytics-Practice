```python
# Compute Daily Revenue

ordersRDD = sc.textFile('/Users/shreyas_rl/Desktop/git/itversity-data/retail_db/orders')
ordersRDD.take(5)
```

```python
mapOrdersRDD = ordersRDD.map(lambda x : x.split(',')[3])
mapOrdersRDD.take(5)
```

```python
# If order Status = Complete/Closed, it is considered in Revenue

revOrdersRDD = ordersRDD.filter(lambda x : x.split(',')[3] in ('COMPLETE','CLOSED'))

revOrdersRDD.take(5)
```

```python
orderItemsRDD = sc.textFile('/Users/shreyas_rl/Desktop/git/itversity-data/retail_db/order_items')
orderItemsRDD.take(5)
```

```python
# For Join Operations, both datasets must be of the type : (K,V) & (K,W)


ordersMap = revOrdersRDD.map(lambda x : func(x))
def func(x):
    temp = x.split(',')
    return((int(temp[0]),temp[1]))

ordersMap.take(5)
```

```python
orderItemsMap = orderItemsRDD.map(lambda x : fun(x))
def fun(x):
    temp = x.split(',')
    return((int(temp[1]),float(temp[4])))

orderItemsMap.take(5)
```

```python
joinedRDD = ordersMap.join(orderItemsMap)
joinedRDD.take(5)
```

```python
joinedRDDMap = joinedRDD.map(lambda x : x[1])
joinedRDDMap.take(5)
```

```python
rbkRDD = joinedRDDMap.reduceByKey(lambda x,y : x+y)
rbkRDD.collect()
```

```python
# Records with no corresponding revenue:
noRevenueRDD = ordersMap.leftOuterJoin(orderItemsMap).filter(lambda x : x[1][1] == None)
noRevenueRDD.collect()
```

```python
sortedRevenueRDD = rbkRDD.sortByKey()
sortedRevenueRDD.collect()
```

```python
# Composite Sorting:

orderItemsRDD.take(5)
```

```python
orderItemsMap = orderItemsRDD.map(lambda x : func(x))
def func(x):
    temp = x.split(',')
    return(((int(temp[1]),float(temp[4])),temp))
orderItemsMap.take(5)
```

```python
# Sort OrderID Ascending, Revenue Ascending
ascSort = orderItemsMap.sortByKey(True)
ascSort.take(5)
```

```python
# Sort OrderID Descending, Revenue Descending
desSort = orderItemsMap.sortByKey(False)
desSort.take(5)
```

```python
# Sort OrderID Ascending, Revenue Descending
desSort = orderItemsRDD.map(lambda x : ((int(x.split(',')[1]),-float(x.split(',')[4])),x)).sortByKey(True)
desSort.take(5)
```
