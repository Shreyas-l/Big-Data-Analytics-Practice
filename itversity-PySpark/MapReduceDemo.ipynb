```python
def read_data(data_path):
    file=open(data_path)
    fileData = file.read()
    fileList = fileData.splitlines()
    return(fileList)
```

```python
data_path = '/Users/shreyas_rl/Desktop/git/itversity-data/retail_db/orders/part-00000'
fileList = read_data(data_path)
```

```python
# Display Complete Orders
```

```python
def complete_orders(fileList):
    completeOrders = []
    for order in fileList:
        order_details = order.split(',')
        if order_details[3] == 'COMPLETE':
            completeOrders.append(order)
    return(completeOrders)
```

```python
completeOrders = complete_orders(fileList)
completeOrders
```

```python
# Display Complete Orders using Lambda Fn (Type 2)
```

```python
def complete_orders(fileList,func):
    completeOrders = []
    for order in fileList:
        order_details = order.split(',')
        if func(order_details):
            completeOrders.append(order)
    return(completeOrders)
```

```python
def findComplete(dataList):
    if dataList[3] == 'COMPLETE':
        return(True)
```

```python
completeOrdersLambda = complete_orders(fileList,lambda x: findComplete(x))

completeOrdersLambda
```

```python
# Python's Filter Function
help(filter)
```

```python
completeOrders = list(filter(lambda x : x.split(',')[3] == 'COMPLETE',fileList))
completeOrders
```

```python
#CALCULATE Total Revenue of Complete Orders using MAP

help(map)
```

```python
revenueComplete = list(map(lambda x : int(x.split(',')[2]),completeOrders))
revenueComplete
```

```python
#Reduce to Sum
import functools as ft
help(ft.reduce)
```

```python
totalRevenue = ft.reduce(lambda x,y : x+y, revenueComplete)
totalRevenue
```
